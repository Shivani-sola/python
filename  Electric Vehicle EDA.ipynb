```python
#Count the number of EVs by State
ev_counts = df.groupby('State').size().reset_index(name='Count')

# Create a choropleth map for the United States
fig = px.choropleth(
    ev_counts,
    locations='State',
    locationmode='USA-states',     # Specifies U.S. state boundaries
    color='Count',
    hover_name='State',
    title='Number of EV Vehicles by State',
    color_continuous_scale=px.colors.sequential.Plasma,
    labels={'Count': 'Number of EV Vehicles'}
)

# Update map to focus on the United States
fig.update_geos(scope="usa")

# Show the map
fig.show()
```

```python
import bar_chart_race as bcr
import pandas as pd
import os
from IPython.display import display, Image

df_cleaned = df_cleaned.copy()
df_cleaned['Model Year'] = df_cleaned['Model Year'].astype(str)
grouped_data = df_cleaned.groupby(['Model Year', 'Make']).size().reset_index(name='Count')
pivoted_data = grouped_data.pivot(index='Model Year', columns='Make', values='Count')
pivoted_data = pivoted_data.fillna(0)

temp_filename = 'temp_EV_racing_bar_plot.gif'
bcr.bar_chart_race(
    df=pivoted_data,
    filename=temp_filename,
    orientation='h',
    sort='desc',
    n_bars=10,
    title='EV Make Count Over the Years',
    filter_column_colors=True,
    period_length=1000
)

display(Image(filename=temp_filename))

os.remove(temp_filename)
```
